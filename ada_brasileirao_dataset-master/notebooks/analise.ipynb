```python
df0   =  pd.read_excel('b-2000/b2000.xlsx')
df1   =  pd.read_excel('b-2001/b2001.xlsx')
df2   =  pd.read_excel('b-2002/b2002.xlsx')
df3   =  pd.read_excel('b-2003/b2003.xlsx')
df4   =  pd.read_excel('b-2004/b2004.xlsx')
df5   =  pd.read_excel('b-2005/b2005.xlsx')
df6   =  pd.read_excel('b-2006/b2006.xlsx')
df7   =  pd.read_excel('b-2007/b2007.xlsx')
df8   =  pd.read_excel('b-2008/b2008.xlsx')
df9   =  pd.read_excel('b-2009/b2009.xlsx')
df10  =  pd.read_excel('b-2010/b2010.xlsx')
df11  =  pd.read_excel('b-2011/b2011.xlsx')
df12  =  pd.read_excel('b-2012/b2012.xlsx')
df13  =  pd.read_excel('b-2013/b2013.xlsx')
df14  =  pd.read_excel('b-2014/b2014.xlsx')
df15  =  pd.read_excel('b-2015/b2015.xlsx')
df16  =  pd.read_excel('b-2016/b2016.xlsx')
df17  =  pd.read_excel('b-2017/b2017.xlsx')
```

```python
def trataPlacar(df):
    #Remove os espaços indesejados
    df         =  df.apply(lambda x: x.str.strip())
    placar     =  df['Placar'].str.split('x')
    df         =  pd.DataFrame(item for item in placar)
    df.columns =  ['p1', 'p2']
    df.index   =  np.arange(0, len(df)+0)
    return df
    
```

```python
dataFrame       =  pd.concat([df0, df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16, df17])
dataFrame.index =  np.arange(0, len(dataFrame)+0)

df         =  trataPlacar(dataFrame)
dataFrame  =  pd.concat([dataFrame, df], axis=1)

dataFrame.drop(['Placar'], inplace=True, axis=1)
dataFrame = dataFrame.apply(lambda x: x.str.strip())
dfCampeonato  =  dataFrame
```

```python
dfCampeonato = dataFrame
dfCampeonato['C1 Estado'] = pd.Series(pd.np.nan, index=dfCampeonato.index)
dfCampeonato['C2 Estado'] = pd.Series(pd.np.nan, index=dfCampeonato.index)
dfCampeonato['Vencedor Estado'] = pd.Series(pd.np.nan, index=dfCampeonato.index)

dfCampeonato            
```

```python
def classificarPorEstadoClube1(dfCampeonato):
    df = dfCampeonato
    for index, row in df.iterrows():
        for key, value in clubes.iteritems():
            c1 = row['Clube 1'].lower()

            if c1 == key.decode('utf-8'):
                df.loc[index, 'C1 Estado'] = value
                break
    return df

def classificarPorEstadoClube2(dfCampeonato):
    df = dfCampeonato
    for index, row in df.iterrows():
        for key, value in clubes.iteritems():
            c2 = row['Clube 2'].lower()

            if c2 == key.decode('utf-8'):
                df.loc[index, 'C2 Estado'] = value
                break
    return df

def classificarPorEstadoVencedor(dfCampeonato):
    df = dfCampeonato
    for index, row in df.iterrows():
        empate = True
        for key, value in clubes.iteritems():
            c2 = row['Vencedor'].lower()

            if c2 == key.decode('utf-8'):
                df.loc[index, 'Vencedor Estado'] = value
                empate = False
                break
        if empate:
            df.loc[index, 'Vencedor Estado'] = 'Empate'
    return df

dfCampeonato = classificarPorEstadoClube1(dfCampeonato)
dfCampeonato = classificarPorEstadoClube2(dfCampeonato)
dfCampeonato = classificarPorEstadoVencedor(dfCampeonato)
```

```python
writer = pd.ExcelWriter('campeonato-brasileiro-full.xlsx')
dfCampeonato.to_excel(writer,'Sheet1', index=False)
writer.save()
#dfCampeonato
```
